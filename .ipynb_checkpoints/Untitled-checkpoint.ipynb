The data in this notebook was downloaded from PhysioNet.  It was originally obtained and used in the Journal Article, Objective estimation of loudness growth in hearing-impaired listeners, by Ikaro Silva and Michael Epstein, printed in Journal of Acoustical Society of America 131, January 2012.  

Three different components are included in each file:
1. The signal presented to the patient
2. The Auditory Brainwave responce (ABR)
3. The Otoacoutic Emission responce (OAE)

```python
# This is an illistration of the three different waveforms included in each file.
record = wfdb.rdrecord('./h1_l_1kHz_50_1') 
wfdb.plot_wfdb(record, title='h1_ABR') 
```

```python
# Rudning display will show all the compoenents of the file in dictionary form
display(record.__dict__)
```

```python
record_1 = record.p_signal
len(record_1)
```

```python
wave = []
for i in record_1:
    wave.append(i[1])
    
plt.plot(wave)
```

```python
oae = []
for i in record_1:
    oae.append(i[2])
    
plt.plot(oae)
```

```python
signal = []
for i in record_1:
    signal.append(i[0])
    
plt.plot(signal)
```

```python
ft = sp.fft.fft(signal)
magnitude = np.absolute(ft)
frequency = np.linspace(0, 48000, len(magnitude))
```

```python
plt.figure(figsize=(18,8))
plt.plot(frequency[:5000], magnitude[:5000])
plt.xlabel('Frequency Hz')
plt.ylabel('Magnitude')
plt.show()
```
